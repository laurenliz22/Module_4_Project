# Business Case

We were hired by New Orleans Real Estate Investment Firm to determine the top 5 "best" zip codes for their company to invest in over the next 5-10 years.  Our client has concerns about both climate change and potential economic crash/fluctuation incidents.  Therefore, the top 5 best zip codes will be those areas that:

- remained more or less stable during events such as Hurricane Katrina and the 2008 economic crash; and 

- forecast increases in prices during the next 5-10 years based on historical data

To look into this request, we've used monthly data from the Zillow Research Page between 4/1/1996 - 4/1/2018.  The data includes all zip codes within the United States as well as pricing data that we'll need to determine the top 5 best zip codes for our client to invest in.  We'll used the historical pricing data to look at how prices have fluctuated over the past 22 years for homes in the New Orleans Metro area.  We'll also looked at the Return on Investment for homes over each month and year, as well as the median and mean prices over the last 5 years, to see if there were any favorable changes.  We will then able to use the time series ARIMA model to fit our data for each of the 5 zip codes to provide us with forecasted prices for the next 5-10 years.

## Change Data to Long Format
Update the format of the data using the melt_data function for our time series analysis

```python
#function to rearrange dataframe to have time as a column with value next to it
def melt_data(df):
    #use pd.melt to rearrange dataframe to long formate for dates
    melted = pd.melt(df, id_vars=['RegionID','RegionName', 'City', 'State', 'Metro', 'CountyName','SizeRank'], var_name='time')
    #set time to datetime format
    melted['time'] = pd.to_datetime(melted['time'], infer_datetime_format=True)
    #drop any NaN's 
    melted = melted.dropna(subset=['value'])
    return melted
```

```python
#run melt function to produce updated dataframe 
df = melt_data(df)
```

## Determine 'Best' 5 Zip Codes

Will create a function to graph each zip code by price x time for each county under consideration to obtain more information to narrow down our search

```python
#define a function to create price to time graphs for each county
def graph(county):
    #create a dataframe for each county in the New Orleans Metro Area
    df_county = df_NO.loc[df_NO.CountyName == county]
    #create a for loop to plot each counties zip code
    for code in list(df_county.RegionName.unique()):
        df_county_code=df_county.loc[df_county.RegionName == code]
        plt.figure(figsize=(3,3))
        plt.plot(df_county_code.time, df_county_code.value, label=code)
        plt.legend()
        plt.title(county)
        plt.ylabel('price')
        plt.xlabel('year')
```

## Five Zip Codes: Set Up CSV Files

- 70122 (Orleans County)
- 70001 (Jefferson County)
- 70068 (Saint John the Baptist County)
- 70037 (Plaquemines County)

```python
#Create dictionary for each zip code 
df_70118.drop(columns=['RegionName', 'CountyName'], axis=1, inplace=True)
df_70118.set_index('time', inplace=True)
df_70118 = df_70118.value.resample('MS').mean()
df_70118 = df_70118.dropna()
df_70118.to_csv('df_70118')

df_70122.drop(columns=['RegionName', 'CountyName'], axis=1, inplace=True)
df_70122.set_index('time', inplace=True)
df_70122 = df_70122.value.resample('MS').mean()
df_70122 = df_70122.dropna()
df_70122.to_csv('df_70130')

df_70001.drop(columns=['RegionName', 'CountyName'], axis=1, inplace=True)
df_70001.set_index('time', inplace=True)
df_70001 = df_70001.value.resample('MS').mean()
df_70001 = df_70001.dropna()
df_70001.to_csv('df_70001')

df_70068.drop(columns=['RegionName', 'CountyName'], axis=1, inplace=True)
df_70068.set_index('time', inplace=True)
df_70068 = df_70068.value.resample('MS').mean()
df_70068 = df_70068.dropna()
df_70068.to_csv('df_70068')

df_70037.drop(columns=['RegionName', 'CountyName'], axis=1, inplace=True)
df_70037.set_index('time', inplace=True)
df_70037 = df_70037.value.resample('MS').mean()
df_70037 = df_70037.dropna()
df_70037.to_csv('df_70037')
```

## Zip Code 70118 (Orleans County)

### Fitting the ARIMA Time Series Model

```python
#create data to train the model using data from 4/1/1996 to 4/1/2015
train_70118 = df_70118[:'2015-04-01']
train_70118.plot(figsize=(15,6))
plt.show()
print(train_70118.head())
```

There is an increasing trend with some seasonality

```python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
pdqs = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
#determine what the best model would be using a for loop
ans = []
for comb in pdq:
    for combs in pdqs:
        try:
            mod = sm.tsa.statespace.SARIMAX(train_70118,
                                            order=comb,
                                            seasonal_order=combs,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            output = mod.fit()
            ans.append([comb, combs, output.aic])
            print('ARIMA {} x {}12 : AIC Calculated ={}'.format(comb, combs, output.aic))
        except:
            continue
```

```python
#print out the best model parameters
ans_df = pd.DataFrame(ans, columns=['pdq', 'pdqs', 'aic'])
ans_df.loc[ans_df['aic'].idxmin()]
```

The output of our code suggests that ARIMA(1, 1, 1)x(1, 1, 1, 12) yields the lowest AIC value of 3171.42. We should therefore consider this to be optimal option out of all the models we have considered.

```python
#Use pdq and pdqs above to fit the data with the ARIMA model
ARIMA_MODEL = sm.tsa.statespace.SARIMAX(train_70118,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

output_train_70118 = ARIMA_MODEL.fit()

print(output_train_70118.summary().tables[1])
```

For our time series, each weight has a p value of 0 (<0.05), except for ar.S.L12 which is slightly above 0.05.  This means this coefficient is less significant to the model then the others.  

```python
#Use plot_diagnostics with results calculated above to make sure 
#our assumptions of normality and correlation hold
output_train_70118.plot_diagnostics(figsize=(15, 18))
plt.show()
```

Findings: 

- Based on the top right kde plot and the bottom left q-q plot, we see that our residuals are close to normally distributed 
- The residuals overtime do display some seasonality in the top left plot, which is confirmed in the bottom right autocorrelation plot showing some of the the time series residuals have larger correlation with the lagged versions of itself.

Using all our data we have p values less than 0.05

#### One Step Ahead Forecasting

We'll use One step ahead forecasting to gain insight into our models predictive power

```python
pred = output_all_70118.get_prediction(start=pd.to_datetime('2013-04-01'), dynamic=False)
pred_conf = pred.conf_int()
```

```python
# Plot real vs predicted values along with confidence interval

rcParams['figure.figsize'] = 15, 6

#Plot observed values
ax = all_70118.plot(label='observed')

#Plot predicted values
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.9)

#Plot the range for confidence intervals
ax.fill_between(pred_conf.index,
                pred_conf.iloc[:, 0],
                pred_conf.iloc[:, 1], color='g', alpha=.5)

#Set axes labels
ax.set_xlabel('Date')
ax.set_ylabel('70118 Levels')
plt.legend()

plt.show()
```

The forecasts align with the true values as seen above, with overall increase trend. 

```python
#Get predictions starting from 04-01-2015 and calculate standard error.
all_70118_forecasted = pred.predicted_mean
all_70118_truth = all_70118['2013-04-01':]
mse = ((all_70118_forecasted - all_70118_truth) ** 2).mean()
print('The Standard Error of our forecasts is {}'.format(round(np.sqrt(mse),4)))
```

#### Dynamic Forecasting

Achieve a deeper insight into model's predictive power using dynamic forecasts.

```python
# Get dynamic predictions with confidence intervals as above.
pred_dynamic = output_all_70118.get_prediction(start=pd.to_datetime('2013-04-01'), dynamic=True, full_results=True)
pred_dynamic_conf = pred_dynamic.conf_int()
```

```python
# Plot the dynamic forecast with confidence intervals.

ax = all_70118.plot(label='observed', figsize=(20, 15))
pred_dynamic.predicted_mean.plot(label='Dynamic Forecast', ax=ax)

ax.fill_between(pred_dynamic_conf.index,
                pred_dynamic_conf.iloc[:, 0],
                pred_dynamic_conf.iloc[:, 1], color='g', alpha=.3)

ax.fill_betweenx(ax.get_ylim(), pd.to_datetime('2013-04-01'), all_70118_forecasted.index[-1], alpha=.1, zorder=-1)

ax.set_xlabel('Date')
ax.set_ylabel('70118 Levels')

plt.legend()
plt.show()
```

Plotting the observed and forecasted values of the time series, we see that the overall forecasts are not very accurate when using dynamic forecasts. All forecasted values (red line) are a off from the ground truth (blue line), within the confidence intervals of our forecast.

```python
# Extract the predicted and true values of our time series
all_70118_forecasted = pred_dynamic.predicted_mean
all_70118_truth = all_70118['2013-04-01':]

# Compute the mean square error
mse = ((all_70118_forecasted - all_70118_truth) ** 2).mean()
print('Standard Error of our forecasts is {}'.format(round(np.sqrt(mse),4)))
```

```python
# Plot future predictions with confidence intervals

ax = all_70118.plot(label='observed', figsize=(20, 15))
prediction.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_conf.index,
                pred_conf.iloc[:, 0],
                pred_conf.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('70118 Levels')

plt.legend()
plt.show()
```

### Fitting the ARIMA Time Series Model

```python
#create data to train the model using data from 4/1/1996 to 4/1/2015
train_70068 = df_70068[:'2015-04-01']
train_70068.plot(figsize=(15,6))
plt.show()
print(train_70068.head())
```

There are some patterns of seasonality in our data with an upward trend

```python
# Define the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
pdqs = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
#determine what the best model would be using a for loop
ans = []
for comb in pdq:
    for combs in pdqs:
        try:
            mod = sm.tsa.statespace.SARIMAX(train_70068,
                                            order=comb,
                                            seasonal_order=combs,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            output = mod.fit()
            ans.append([comb, combs, output.aic])
            print('ARIMA {} x {}12 : AIC Calculated ={}'.format(comb, combs, output.aic))
        except:
            continue
```

```python
#print out the best model
ans_df = pd.DataFrame(ans, columns=['pdq', 'pdqs', 'aic'])
ans_df.loc[ans_df['aic'].idxmin()]
```

The output of our code suggests that ARIMA(1, 1, 1)x(1, 1, 1, 12) yields the lowest AIC value of 2993.15. We should therefore consider this to be optimal option out of all the models we have considered.

```python
#Use pdq and pdqs above to fit the data with the ARIMA model
ARIMA_MODEL = sm.tsa.statespace.SARIMAX(train_70068,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

output_train_70068 = ARIMA_MODEL.fit()

print(output_train_70068.summary().tables[1])
```

For our time series, we see some weights have a pvalue above 0.05 meaning they are less significant for our model

```python
#Use plot_diagnostics with results calculated above to make sure 
#our assumptions of normality and correlation hold
output_train_70068.plot_diagnostics(figsize=(15, 18))
plt.show()
```

Findings: 

- Based on the top right kde plot and the bottom left q-q plot, we see that our residuals are close to normally distributed 
- The residuals overtime do display some seasonality in the top left plot, which is confirmed in the bottom right autocorrelation plot showing some of the the time series residuals have larger correlation with the lagged versions of itself.
